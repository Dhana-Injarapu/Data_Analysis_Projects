### Let's explore the numerical columns

```python

df['Installs'].unique()
```

"installs" is in object type ,we need to convert in integer type

```python

df['Installs']=df['Installs'].str.split('+').str[0]
```

```python
df['Installs'].unique()
```

Replace "," with nothing

```python
df['Installs'].replace(",","",regex=True,inplace=True)
df['Installs']=df['Installs'].astype('int64')
```

```python
df['Installs'].unique()
```

```python
df['Size'].unique()
```

Here size is in mb,gb,kb convert all data into mb

```python
#df['Size']=df['Size'].apply(lambda x:float(x))
```

```python
#df['Size']=df['Size'].apply(lambda x: float(str(x).replace('M','')) if 'M' in str(x) else x)
```

 assume ',' is incorrectly entered '.' so replace ',' with'.'

```python
df['Size']=df['Size'].str.replace(",",".")
```

```python
df['Size']=df['Size'].apply(lambda x: float(str(x).replace('M','')) if 'M' in str(x) else x)
```

```python
df['Size']=df['Size'].apply(lambda x:float(str(x).replace('G',''))*1000 if 'G' in str(x) else x )
```

```python
df['Size']=df['Size'].apply(lambda x:float(str(x).replace('k',''))/1000 if 'k' in str(x) else x)
```

```python
#df['Size']=df['Size'].apply(lambda x:float(x))
```

here size is varies with device ,lets assume it as 0

```python
df['Size']=df['Size'].apply(lambda x:str(x).replace('Varies with device','0') if 'Varies with device' in str(x) else x)
```

```python
df['Size']=df['Size'].apply(lambda x:float(x))
```

There is a high variance in the number of installs, we need to reduce it so we can use a log value for this column,
otherwise it would be unable to see the data when we visualize

```python
categoty_type_installs = df.groupby(['Category','Type'])[['Installs']].sum().reset_index()
categoty_type_installs
```

```python
categoty_type_installs['log_Installs'] = np.log10(categoty_type_installs['Installs'])
```

```python
plt.figure(figsize=(18,9))
plt.xticks(rotation=65,fontsize=9)
plt.xlabel("Category")
plt.ylabel("Installs")
plt.title("Number of installed Apps type wise according to Category")
sns.barplot(x = 'Category', y = 'log_Installs', hue='Type', data=categoty_type_installs)
```
